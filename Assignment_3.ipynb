```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("https://zenodo.org/record/11394760/files/athlete_events-1.csv?download=1")

# Drop rows with missing values in important columns
df.dropna(subset=["Age", "Height", "Weight", "Sex", "Sport", "Medal", "Year"], inplace=True)
```

```python
sns.histplot(df["Age"], bins=30, kde=True, color="skyblue")
plt.title("Age Distribution of Athletes")
plt.show()
```

```python
sns.boxplot(x="Medal", y="Age", data=df, palette="pastel")
plt.title("Age by Medal Type")
plt.show()
```

```python
sns.scatterplot(data=df, x="Height", y="Weight", hue="Sex", alpha=0.5)
plt.title("Height vs Weight of Athletes")
plt.show()
```

```python
df.groupby("Year")["ID"].nunique().plot(marker='o')
plt.title("Number of Athletes Over the Years")
plt.xlabel("Year"); plt.ylabel("Number of Athletes")
plt.grid(True)
plt.show()
```

```python
df["Sex"].value_counts().plot.pie(autopct="%1.1f%%", colors=["#ff9999", "#66b3ff"])
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
```

```python
top_sports = df[df["Medal"].notna()]["Sport"].value_counts().head(10)
sns.barplot(x=top_sports.values, y=top_sports.index, palette="viridis")
plt.title("Top 10 Sports by Medal Count")
plt.xlabel("Number of Medals")
plt.show()
```

```python
sns.countplot(data=df[df["Medal"].notna()], x="Sex", hue="Medal", palette="Set2")
plt.title("Medal Distribution by Gender")
plt.show()
```

```python
numeric_df = df[["Age", "Height", "Weight"]]
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Numerical Features")
plt.show()
```

```python
g = sns.FacetGrid(df, col="Sex")
g.map(sns.histplot, "Age", bins=30)
plt.suptitle("Age Distribution by Gender", y=1.05)
plt.show()
```
